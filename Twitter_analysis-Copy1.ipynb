```python
import csv
reader=csv.reader(open('spicejet_new.csv','rb'))
tweets=set()
for row in reader:
    tweet=[]
    tweet.append(list(row[0].split(';'))[4])
    #print tweet
    tweets.update(set(tweet))
for tweet in tweets:
    tweet=tweet.split()
    for word in tweet:
        if word[0]=='"':
            tweet[tweet.index(word)]=word[1:]
            print '\n\n\n'
        if word[0]=='@':
            tweet[tweet.index(word)]='USERNAME'
            print '\n\n\n'
    print tweet
```
